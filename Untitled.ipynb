```python
import pickle
import numpy as np
from rdkit.Chem.Draw import rdMolDraw2D
from rdkit import Chem
from IPython.display import SVG
from rdkit.Chem.Draw import MolToImage
from rdkit.Chem import Draw

def idxForSignificants(th_, aWeight):
    retlist = []
    for idx, ele in enumerate(aWeight):
        if ele >= th_:
            retlist.append(idx)
    return retlist

def idxSeparate(sigIdx, currSmiList):
    atomIdx, bondIdx = [], []
    atomCnt = 0
    bondCnt = 0
    for idx, ele in enumerate(currSmiList):
        if ele.isalpha():
            if idx in sigIdx:
                atomIdx.append(atomCnt)
            atomCnt += 1
        else:
            if idx in sigIdx:
                bondIdx.append(bondCnt)
            bondCnt += 1
    return atomIdx, bondIdx

def idxSeparate2(sigIdx, currSmiList):
    atomIdx, bondIdx = [0 for i in range(len(currSmiList))], [0 for i in range(len(currSmiList))]
    
    for idx, ele in enumerate(currSmiList):
        if idx in sigIdx:
            if ele.isalpha():
                atomIdx[idx] = 1
            else:
                bondIdx[idx] = 1
    return atomIdx, bondIdx

#attnweights, teststr = pickle.load( open( "../save_weights", "rb" ) )
convweights, teststr = pickle.load( open( "../save_weights", "rb" ) )
```

```python
# conv_weights.shape == (batch_size, seq_len, self.d_model)
print(len(convweights))
layer1 = np.squeeze(convweights)#layer1 shape = (whole_size, seq_len, self.d_model)
layer1_sum = np.sum(layer1, axis=2)#layer1T_sum shape = (whole_size, seq_len)
#print(layer1T_sum[0], max(layer1T_sum[0]))
allMols = []
allAtomIdx = []
allBondIdx = []
for i, ele in enumerate(layer1_sum):
    th_Until7thW = sorted(ele)[-7]
    sigIdx = idxForSignificants(th_Until7thW, ele)
    atomIdx, bondIdx = idxSeparate(sigIdx, teststr[i])
    smi = "".join(teststr[i])
    mol = Chem.MolFromSmiles(smi)
    allMols.append(mol)
    allAtomIdx.append(atomIdx)
    allBondIdx.append(bondIdx)

#img = MolToImage(mol, size=(600, 600), fitImage=False)#, highlightAtoms=atomIdx, highlightBonds=bondIdx)   
#img
#print(smi)
img = Draw.MolsToGridImage( allMols, molsPerRow=2, highlightAtomLists=allAtomIdx, highlightBondLists=allBondIdx, useSVG=True, subImgSize=(400, 400), maxMols=100)
img
```
